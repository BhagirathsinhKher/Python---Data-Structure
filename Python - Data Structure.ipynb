**Q.4 Describe how dictionaries store data.**

**Ans.**
- Dictionaries in Python use a hash table to store the key-value pairs. Here's how it works:
>- Hashing: When a key is added to the dictionary, Python calculates a hash value for the key using a hash function. This hash value determines the index where the key-value pair will be stored in the underlying array (or hash table).
>- Key-Value Mapping: Each key is mapped to a specific value. The hash function ensures that each key points to the correct memory location where the value is stored.
>- Handling Collisions: In cases where two keys generate the same hash value (a collision), Python uses a method called open addressing or separate chaining to resolve the conflict and store both key-value pairs.

**Q.8 What is a hash table, and how does it relate to dictionaries in Python?**

**Ans.**

- A hash table is a data structure used to store key-value pairs. It provides a fast way to look up values associated with specific keys, making it highly efficient for certain operations. In Python, the dictionary (dict) is implemented using a hash table.
- In Python, dictionaries are implemented using hash tables, which means they store key-value pairs efficiently. Here's how it works with Python dictionaries:
>- Keys Must Be Hashable: Since the keys in a dictionary are hashed to find their location in memory, the keys must be immutable and hashable. Examples of hashable types include strings, numbers, and tuples (if they contain only immutable elements). Lists, on the other hand, are not hashable because they are mutable.
>- Fast Lookups: Because Python dictionaries use hash tables, accessing a value by its key is done in constant time on average.
>- Collisions: Pythons dictionary implementation handles collisions internally using efficient strategies like open addressing or chaining. However, in most cases, Pythons hash table implementation avoids collisions, or it handles them efficiently when they do occur.
>- Uniqueness of Keys: In a dictionary, each key must be unique. If you try to add a duplicate key, the existing key-value pair will be overwritten with the new one.

**Q.13 How do sets handle duplicate values in Python?**

**Ans.**
- In Python, sets automatically eliminate duplicate values. When you attempt to add duplicate elements to a set, Python ensures only one instance of each unique value is stored. This behavior is built into the set's data structure, which uses a hash table to manage its elements.
- Hash Table-Based Storage: Sets rely on hash values to store elements. If a new elements hash value matches that of an existing element, it is recognized as a duplicate and not added.
- Automatic Deduplication: Python removes duplicates silently, ensuring that sets only contain unique values without requiring manual intervention.

**Q.17 Describe the time complexity of accessing elements in a dictionary.**

**Ans.**

- The time complexity of accessing elements in a dictionary in Python is O(1) on average. This means that accessing a value by its key takes constant time, regardless of the size of the dictionary.
- This efficiency is due to the underlying hash table implementation of dictionaries.
- *How It Works:*
>- *Hashing:* When you access a key in a dictionary, Python uses a hash function to compute the hash value of the key. This hash value is used to find the location (bucket) where the corresponding value is stored in the hash table.
>- *Direct Access:* Once the correct bucket is found, the value associated with the key is directly retrieved, making the lookup operation very fast.

**Q.19 Why are dictionaries considered unordered, and how does that affect data retrieval?**

**Ans.**

- In Python, dictionaries are considered unordered because the elements in a dictionary are stored using a hash table that does not guarantee any specific order of the keys. This means that when you iterate over the keys or values of a dictionary, the order in which the elements are returned may not be the same as the order in which they were added.
- The primary goal of dictionaries is fast lookups (average O(1) time complexity) based on keys, not maintaining order. Hashing allows for quick access to values by their corresponding keys, but at the cost of losing the order of insertion.
- *How Does This Affect Data Retrieval?*
- No Indexing or Slicing:
>- Unlike lists, dictionaries do not support indexing or slicing because they are unordered. You cannot retrieve an element based on its position in the dictionary. Instead, you must access elements via their keys.
- Retrieving Data:
>- To retrieve a value from a dictionary, you must use the key directly, which is fast due to the hashing mechanism. However, you cannot assume any order when doing so.
- Data retrieval is fast (O(1) on average) based on the key, but you cannot rely on the order.
---

**Q.20 Explain the difference between a list and a dictionary in terms of data retrieval.**

**Ans.**

- *Lists:* Data is retrieved by index. Each element in a list is associated with an integer index that represents its position in the list.
- Retrieval by index has a time complexity of O(1) because the position is known directly.
- Indices are sequential, starting from 0, so the retrieval depends on the position of the element.
- *Dictionaries:* Data is retrieved by key. Each element (a key-value pair) is stored in a way that the key is hashed to determine its location in a hash table.
- Retrieval by key has a time complexity of O(1) on average, regardless of the size of the dictionary.
- Keys can be any hashable object, providing flexibility to retrieve elements by meaningful identifiers rather than numerical indices.
